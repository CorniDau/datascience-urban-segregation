### Datenlage:
Das Informationsportal der Stadt Dortmund liefert sehr gut aufbereitete und vor allem einheitliche (Stadtbezirke = Wahlbezirke) Daten für die jeweiligen Stadtbezirke im Jahr 2017: https://opendata.dortmund.de/Informationsportal/.

Die Wahlergebnisse wurden auf Anfrage vom Statistikamt der Stadt Dortmund bereit gestellt. 

```python
# Pandas Bibliothek als pd importieren:
import pandas as pd

# .csv Datei einlesen, welche im gleichen Verzeichnis wie das Notebook liegt;
# Die Stadtbezirke (erste Spalte) werden als Index genutzt;
# Da es beim Import der Daten nach Standardwerten eine Fehlermeldung gab,
# haben wir herausgefunden, dass im deutschen Raum der Separator ; verwendet wird,
# um Kommazahlen zu ermöglichen. Somit muss auch beim Import auf den entsprechenden Separator eingegangen werden;
# Datensatz wird in der Variablen "dortmund" gespeichert;
dortmund = pd.read_csv('Datentabellen_GruppeB.csv', index_col=0, sep=';')
```

```python
# Um dem Problem vorzubeugen, dass der Bezeichner der ersten Spalte als erste Zeile
# gelesen wird kommen die folgenden Zeilen zum Einsatz:
dortmund.columns.name = dortmund.index.name
dortmund.index.name = None
print (dortmund)
```
